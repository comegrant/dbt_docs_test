```python
from sklearn.ensemble import RandomForestClassifier
```

```python
class ModelConfig:
    def __init__(self):
        self.classifier_mapping = {
            "AMK": {
                "has_chefs_favorite_taxonomy": self.get_random_forest_classifier(
                    max_depth=14, min_samples_leaf=2, min_samples_split=10, n_estimators=231, random_state=42
                ),
                "has_family_friendly_taxonomy": self.get_random_forest_classifier(
                    max_depth=14, min_samples_leaf=2, min_samples_split=10, n_estimators=231, random_state=42
                ),
            },
            # RT and AMK
        }

        self.model_uri_mapping = {
            "AMK": {
                "has_chefs_favorite_taxonomy": "runs:/123/pyfunc_packaged_model",
                "has_family_friendly_taxonomy": "runs:/456/pyfunc_packaged_model",
            },
            # RT and AMK
        }

    @staticmethod
    def get_random_forest_classifier(**params):
        return RandomForestClassifier(**params)


    def classifier(self, company, target):
        if company not in self.classifier_mapping or target not in self.classifier_mapping[company]:
            raise ValueError(f"No classifier found for company '{company}' and target '{target}'")
        return self.classifier_mapping[company][target]
    
    def model_uri(self, company, target):
        if company not in self.model_uri_mapping or target not in self.model_uri_mapping[company]:
            raise ValueError(f"No model uri found for company '{company}' and target '{target}'")
        return self.model_uri_mapping[company][target]
```
